Determine the average popularity for each movie subject category and identify the top 3 most popular subjects.

```python
# Group by 'Subject' using groupby() method.
# Calculate the average of Popularity using mean() method.

subject_category = data.groupby('Subject')['Popularity'].mean()
#Use sort_values() method to sort the Subjects by average popularity values in descending order.
# Use head() method to show the top 3 most popular subjects.

subject_category_top_3 = subject_category.sort_values(ascending = False).head(3)
print('Group By Movie Subject:')
print(subject_category)
print('\nTop 3 Movie Subject by Populairty:')
print(subject_category_top_3)
```

```python
#Use dropna() method to remove nulls in Length, Subject, Actor, Popularity columns.
#Use replace() method to replace the nulls with ‘Unknown’ in Actress and Director columns.
data.dropna(subset = ['Length', 'Subject', 'Actor', 'Popularity'])
data[['Actress','Director']] =data[['Actress','Director']].replace('\\N','Unknown')
data
```

another method


```python
#Use replace() method to replace the nulls with ‘Unknown’ in Actress and Director columns.
data.dropna(inplace=True)
data['Actress'] = data['Actress'].replace('\\N', 'Unknown')
data['Director'] = data['Director'].replace('\\N', 'Unknown')
data
```

Explore the trend of average movie lengths over the years between 1980 and 2000 with award status breakdown.

```python
#Filter the movies based on the 'Year' column from 1980 to 2000
avg_length_1980_2000 = data[(data['Year'] >=1980 ) & (data['Year'] <=2000)]

avg_length_1980_2000
```

```python
#Group the data by 'Year' and 'Awards' and calculate the average 'Length'.
year_by_award = avg_length_1980_2000.groupby(['Year' , 'Awards']).Length.mean().reset_index()

# OR
# Filtering movies from 1980 to 2000 and calculating the average length
data_80s_00s = data[(data['Year'] >= 1980) & (data['Year'] <= 2000)]
data_80s_00s_awards = data_80s_00s.groupby(['Year','Awards'])['Length'].mean().reset_index()

#avg_length = data['Length'].mean()
year_by_award
#data_80s_00s_awards
```
